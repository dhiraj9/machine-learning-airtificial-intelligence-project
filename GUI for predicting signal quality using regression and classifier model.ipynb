```python
import tkinter as t
from tkinter import ttk
import tensorflow.keras as k
w=t.Tk()
w.title('Neural Networks GUI - Great Learning')
n=ttk.Label(w,text='Step 1: File Name')
n.grid(row=0,column=0,sticky=t.W)
v=t.StringVar()
a=ttk.Entry(w,width=16,textvariable=v)
a.grid(row=0,column=1)
import pandas as p
import re
def i():
    global d
    b=v.get()
    db=re.findall('\..*',b)
    d=p.read_csv(b)
    e=ttk.Entry(w,width=16)
    e.grid(row=0,column=3)
    e.insert(1,str('Done'))
f=ttk.Button(w,text='Import Data',command=i)
f.grid(row=0,column=2)
g=ttk.Label(w,text='Step 2: Target Column')
g.grid(row=1,column=0,sticky=t.W)
h=t.StringVar()
j=ttk.Entry(w,width=16,textvariable=h)
j.grid(row=1,column=1)
from tensorflow.keras.optimizers import Adam
def ak():
    global l
    m=h.get()
    l=d[m]
    n=ttk.Entry(w,width=16)
    n.grid(row=1,column=3)
    n.insert(1,str('Found'))
o=ttk.Button(w,text='Select Target',command=ak)
o.grid(row=1,column=2)
q=ttk.Label(w,text='Step 3: Neural Network Regressor')
q.grid(row=2,column=0,sticky=t.W)
ar=ttk.Label(w,text='Regression')
ar.grid(row=3,column=0)
from sklearn.model_selection import train_test_split
def s():
    global m
    for i in d.columns[:-1]:
        q1=d[i].quantile(.25)
        q3=d[i].quantile(.75)
        iq=q3-q1
        d.loc[(d[i]<(q1-1.5*iq))|(d[i]>(q3+1.5*iq)),i]=d[i].median()
    x=d.drop(['Signal_Strength'],axis=1)
    xt,xte,yt,yte=train_test_split(x,d['Signal_Strength'],test_size=.3,random_state=0)
    import random
    random.seed(0)
    m=k.models.Sequential([k.layers.BatchNormalization(input_shape=(11,)),k.layers.Dense(512,activation='relu',kernel_initializer='he_normal'),k.layers.BatchNormalization(),k.layers.Dense(256,activation='relu',kernel_initializer='he_normal'),k.layers.BatchNormalization(),k.layers.Dense(128,activation='relu',kernel_initializer='he_normal'),k.layers.BatchNormalization() ,k.layers.Dense(64,activation='relu',kernel_initializer='he_normal'),k.layers.BatchNormalization() ,k.layers.Dense(1)])
    from keras import backend as K
    def r(y,yp):
        aaa=K.sum(K.square(y-yp))
        aab=K.sum(K.square(y-K.mean(y)))
        return (1-aaa/(aab+.0000001))
    m.compile(optimizer=Adam(learning_rate=.0001),loss='mse',metrics=[r])
    m.fit(xt,yt,validation_data=(xte,yte),epochs=1000,batch_size=200)
    at=ttk.Entry(w,width=18)
    at.grid(row=3,column=2)
    at.insert(1,str('Network Trained'))
u=ttk.Button(w,text='Train',command=s)
u.grid(row=3,column=1)
av=ttk.Label(w,text='Pickle')
av.grid(row=4,column=0)
def aw():
    m.save('m.h5')
    z=ttk.Entry(w,width=18)
    z.grid(row=4,column=2)
    z.insert(1,str('Saved model to disk'))
aa=ttk.Button(w,text='Run',command=aw)
aa.grid(row=4,column=1)
ab=ttk.Label(w,text='Step 3: Neural Network Classifier')
ab.grid(row=5,column=0,sticky=t.W)
ac=ttk.Label(w,text='Classifier')
ac.grid(row=5,column=0)
def ad():
    global m
    x=d.drop(['Signal_Strength'],axis=1)
    xt,xte,yt,yte=train_test_split(x,d['Signal_Strength'],test_size=.3,random_state=0)
    import random
    random.seed(0)
    m=k.models.Sequential([k.layers.BatchNormalization(input_shape=(11,)),k.layers.Dense(200,activation='relu'),k.layers.Dense(100,activation='relu'),k.layers.Dense(60,activation='relu'),k.layers.Dense(30,activation='relu'),k.layers.Dense(6,activation='softmax')])
    m.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
    yt=yt-3
    yte=yte-3
    yt=k.utils.to_categorical(yt,num_classes=6)
    yte=k.utils.to_categorical(yte,num_classes=6)
    m.fit(xt,yt,epochs=100,batch_size=64,validation_data=(xte,yte))
    ae=ttk.Entry(w,width=18)
    ae.grid(row=5,column=2)
    ae.insert(1,str('Network Trained'))
af=ttk.Button(w,text='Train',command=ad)
af.grid(row=5,column=1)
ag=ttk.Label(w,text='Pickle')
ag.grid(row=6,column=0)
def ah():
    m.save('mc.h5')
    ai=ttk.Entry(w,width=18)
    ai.grid(row=6,column=2)
    ai.insert(1,str('Saved model to disk'))
aj=ttk.Button(w,text='Run',command=ah)
aj.grid(row=6,column=1)
w.mainloop()
```
