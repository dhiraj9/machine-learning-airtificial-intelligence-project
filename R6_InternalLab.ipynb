# Q5. Build a User-Based Collaborative Filtering model



1.   Consider users who have rated atleast 100 books
2.   Transform data to surprise format
3.   Assure that only user, book and rating columns are present in the new dataset in surprise format



```python
ratings_new=ratings_new.groupby('userID').filter(lambda x:len(x)>99)
ratings_new
```

```python
#2 Transform data to surprise format
from surprise import Dataset,Reader
from surprise.model_selection import cross_validate
from surprise import NormalPredictor


reader = Reader(rating_scale=(1, 10))
```

```python
ratings_new.userID=ratings_new.userID.astype(str)
ratings_new.ISBN=ratings_new.ISBN.astype(str)
da=Dataset.load_from_df(ratings_new[['userID','ISBN','bookRating']],reader)
da
```

# Q6. Apply SVD for user based filtering


1.   Split the data into train and test
2.   Train model on SVD
3.   Test Accuracy of the model




```python
from surprise.model_selection import train_test_split
t,te=train_test_split(da,test_size=.25,random_state=11)
from surprise import SVD,accuracy
s=SVD(n_factors=50,biased=False)
s.fit(t)
accuracy.rmse(s.test(te))
```
