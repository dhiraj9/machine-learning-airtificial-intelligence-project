```python
from surprise import SVD,accuracy,Dataset,KNNWithMeans,Reader
from surprise.model_selection import cross_validate,train_test_split
r=Reader(rating_scale=(1,10))
dat=Dataset.load_from_df(d[['author','product','score']],r)
t,te=train_test_split(dat,test_size=.25)
s=SVD()
s.fit(t)
accuracy.rmse(s.test(te))
```

```python
s.test(te)
```

```python
from collections import defaultdict
de=defaultdict(list)
defa=defaultdict(list)
for uid,iid,r,est,_ in s.test(te):
    de[uid].append((iid,est))
    defa[uid].append((iid,est))
for uid,u in de.items():
    u.sort(key=lambda x:x[1],reverse=True)
    de[uid]=u[:5][1][0]
    defa[uid]=u[1][1]
defa  
```
